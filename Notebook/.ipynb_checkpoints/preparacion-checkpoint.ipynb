```python
def filtro(habitaciones,baños=None,parqueaderos,area):
    consulta = f'SELECT * FROM property WHERE habitaciones = {habitaciones} AND baños = {baños} AND parqueaderos = {parqueaderos} AND area_construida > {area};'
    data_sql = pd.read_sql_query(consulta, conn)
    return data_sql
```

```python
def consulta_filtrada(conn, tabla, columnas="*", condiciones=None):
    """
    Realiza una consulta filtrada en una base de datos SQLite.

    Args:
        conn (sqlite3.Connection): Conexión a la base de datos.
        tabla (str): Nombre de la tabla a consultar.
        columnas (str o list): Columnas a seleccionar, puede ser "*" o una lista de nombres de columnas.
        condiciones (str, opcional): Condiciones para filtrar las filas (e.g., "Precio > 500000").
    
    Returns:
        pd.DataFrame: Resultado de la consulta en formato DataFrame.
    """
    # Convertir la lista de columnas a una cadena separada por comas
    if isinstance(columnas, list):
        columnas = ", ".join(columnas)
    
    # Construir la consulta
    consulta = f"SELECT {columnas} FROM {tabla}"
    if condiciones:
        consulta += f" WHERE {condiciones}"
    
    # Ejecutar la consulta y devolver un DataFrame
    try:
        df = pd.read_sql_query(consulta, conn)
        return df
    except Exception as e:
        print(f"Error al realizar la consulta: {e}")
        return None
```
